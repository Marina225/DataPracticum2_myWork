```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import cluster

face = pd.read_csv("face.csv", index_col = False)
# Da dobimo pravilen format tabele
#face = face.iloc[:,1:3]
face
```

```python
plt.scatter(face.iloc[:,0], face.iloc[:,1], c = "blue", marker = 'o', s = 2)
plt.show()
plt.scatter(face.iloc[:,1], face.iloc[:,2], c = "blue", marker = 'o', s = 2)
plt.show()
```

```python
dist =[]
for i in range(1,10):
    km = cluster.KMeans(n_clusters = i, init = "k-means++")
    km.fit(face)
    dist.append(km.inertia_)
plt.plot(range(1,10), dist, "o-")
plt.show()
```

```python
# Optimalno število klastrov je med 2 in 3. Jaz sem izbrala število 3, saj se glede na podatke zdi da so res 3 večje gruče.

# Razdelitev podatke v klastre
km = cluster.KMeans(n_clusters = 3, init="k-means++")
face_c = km.fit_predict(face.iloc[:,[1,2]])

print(face_c)
```

```python
plt.scatter(face.iloc[face_c == 0, 1], face.iloc[face_c == 0, 2], c = "green", marker = "x", label = "cluster 1")
plt.scatter(face.iloc[face_c == 1, 1], face.iloc[face_c == 1, 2], c = "blue", marker = "o", label = "cluster 2")
plt.scatter(face.iloc[face_c == 2, 1], face.iloc[face_c == 2, 2], c = "red", marker = "v", label = "cluster 3")
plt.legend()
plt.grid()
plt.show()
```

```python
# Po sliki jasno vidimo 3 klastre, ki lepo ločijo oči, nos in usta na obrazu.
```

```python
# Če bi odstranili prvo index kolono, bi po elbow metodi dobili 2 optimalna klastra. 
```

```python
# Da dobimo pravilen format tabele
face_2 = face.iloc[:,1:3]
```

```python
plt.scatter(face_2.iloc[:,0], face_2.iloc[:,1], c = "blue", marker = 'o', s = 2)
plt.show()
```

```python
dist =[]
for i in range(1,10):
    km = cluster.KMeans(n_clusters = i, init = "k-means++")
    km.fit(face_2)
    dist.append(km.inertia_)
plt.plot(range(1,10), dist, "o-")
plt.show()
```

```python
# Optimalno število klastrov je med 2 in 3. Jaz sem izbrala število 3, saj se glede na podatke zdi da so res 3 večje gruče.

# Razdelitev podatke v klastre
km = cluster.KMeans(n_clusters = 2, init="k-means++")
face_c2 = km.fit_predict(face_2.iloc[:,[0,1]])

print(face_c2)
```

```python
# Končni rezultat zgleda takole: 

plt.scatter(face_2.iloc[face_c2 == 0, 0], face_2.iloc[face_c2 == 0, 1], c = "green", marker = "x", label = "cluster 1")
plt.scatter(face_2.iloc[face_c2 == 1, 0], face_2.iloc[face_c2 == 1, 1], c = "blue", marker = "o", label = "cluster 2")
plt.legend()
plt.grid()
plt.show()
```
