dataset: face.csv
 
• perform and visualize clustering

• initial inspection (scatterplots)

• choice of number of clusters (elbow method)

• visualize clusters only with the second and third variable (only to variables are available though)


Code:

```python
# imports
import pandas as pd
from sklearn import cluster
from matplotlib import pyplot as plt
```

```python
# import dataset
df = pd.read_csv("face.csv")
df = df[["V1", "V2"]]
df
```

```python
#visualisation of data
plt.scatter(df.iloc[:,0], df.iloc[:,1], c = "blue", marker = 'o', s = 2)
plt.show()
```

```python
# elbow method to asses the optimal number of clusters
distortions = []
for i in range(1,11):
    km = cluster.KMeans(n_clusters = i, init ="k-means++", n_init=10, max_iter=300, random_state=0)
    km.fit(df)
    distortions.append(km.inertia_)
plt.plot(range(1,11), distortions, "o-")
plt.show()

# the optimal number of clusters is somewhere between 2 and 4 based on the chart of elbow method
```

```python
# prediction - using 2 clusters
print("prediction - using 2 clusters")
k_means = cluster.KMeans(n_clusters=2,init="k-means++")
k_means.fit(df)
prediction2 = k_means.predict(df)
print(prediction2)

# prediction - using  clusters
print("prediction - using 3 clusters")
k_means = cluster.KMeans(n_clusters=3,init="k-means++")
k_means.fit(df)
prediction3 = k_means.predict(df)
print(prediction3)

# prediction - using 4 clusters
print("prediction - using 4 clusters")
k_means = cluster.KMeans(n_clusters=4,init="k-means++")
k_means.fit(df)
prediction4 = k_means.predict(df)
print(prediction4)
```

```python
## plotting 2, 3 and 4 k clusteringt
print("plot - using 2 clusters")
plt.scatter(df.iloc[prediction2 ==0, 0],df.iloc[prediction2 ==0, 1], c = "green", marker = "x", label = "cluster 1")
plt.scatter(df.iloc[prediction2 ==1, 0],df.iloc[prediction2 ==1, 1], c = "blue", marker = "o", label = "cluster 2")
plt.legend()
plt.grid()
plt.show()

print("plot - using 3 clusters")
plt.scatter(df.iloc[prediction3 ==0, 0],df.iloc[prediction3 ==0, 1], c = "green", marker = "x", label = "cluster 1")
plt.scatter(df.iloc[prediction3 ==1, 0],df.iloc[prediction3 ==1, 1], c = "blue", marker = "o", label = "cluster 2")
plt.scatter(df.iloc[prediction3 ==2, 0],df.iloc[prediction3 ==2, 1], c = "orange", marker = "p", label = "cluster 3")
plt.legend()
plt.grid()
plt.show()

print("plot - using 4 clusters")
plt.scatter(df.iloc[prediction4 ==0, 0],df.iloc[prediction4 ==0, 1], c = "green", marker = "x", label = "cluster 1")
plt.scatter(df.iloc[prediction4 ==1, 0],df.iloc[prediction4 ==1, 1], c = "blue", marker = "o", label = "cluster 2")
plt.scatter(df.iloc[prediction4 ==2, 0],df.iloc[prediction4 ==2, 1], c = "orange", marker = "p", label = "cluster 3")
plt.scatter(df.iloc[prediction4 ==3, 0],df.iloc[prediction4 ==3, 1], c = "red", marker = "s", label = "cluster 4")
plt.legend()
plt.grid()
plt.show()
```

Based on the plots and elbow method. I would choose 2 clusters as the optimal number of clusters, that separates eyes from the nose and the mouth.
